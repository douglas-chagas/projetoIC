```python
df_2017 = pd.read_excel("CAMPINA DA LAGOA.xlsx", sheet_name="2017")

display(df_2017)

# Reduzindo dados com groupby
df_diario_2017 = df_2017.groupby(df_2017['Data']).agg({
'Temp. Max. (C)': 'max',
'Temp. Min. (C)': 'min',
'Umi. Max. (%)': 'max',
'Umi. Min. (%)': 'min',
'Vel. Vento (m/s)': 'mean',
'Radiacao (KJ/m²)': 'sum',
'Chuva (mm)': 'sum',
'Pressao Max. (hPa)': 'max',
'Pressao Min. (hPa)': 'min'
})

display(df_diario_2017)
```
